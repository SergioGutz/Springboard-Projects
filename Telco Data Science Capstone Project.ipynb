# Capstone Mini-Project: Data Wrangling

```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset 
data = pd.read_csv('telco_customer_data.csv')
```

```python
# Review dataset composition and inspect for any missing values
data.info()
```

```python
# Check a sample of the data to inspect the numeric and object columns further
data.head()
```

```python
# Convert Total Charges column from an object to numeric column
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'],  errors='coerce')

# Check for outliers in the four numeric columns 
# None of max values for each column is 3 or more standard deviations away from the mean
data_num = data.select_dtypes(include=['number'])
data_num.describe()
```

```python
# Check all category columns and isolate them for further conversion into binary columns
data_cat = data.iloc[:,1:].select_dtypes(include=[object])
data_cat.head()
```

```python
# Use get dummies to convert the categorical columns into binary columns, dropping the first possible outcome per category
dummies = pd.get_dummies(data_cat, prefix_sep='_', drop_first=True)
dummies.head()
```

```python
# Combine the customer ID column with the numeric and new binary category columns
df = pd.concat([data['customerID'], data_num, dummies], axis=1)
df.head()
```
